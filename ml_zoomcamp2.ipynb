# ML vs Rule-Based Systems

Here, we will compare the traditional rule based system vs the ML based approach using the example of spam detection.

**i.e.** we want to build a **classifier** that classifies emails as spam or not spam.

So we come up with rules that determine if email is spam or not

But then a new email arrives without these rules being true but it is still marked spam by users.  So we add another rule based on that new spam mail's characteristic. And this happens again, and we do this over and over. **NEVER ending process**. more code code and more code.


#### ML approach:
get data

define and calculate features

train and use the model

We can get data generated by users marking all emails they get as spam or not.


We actually find that the rules we developed in the rule based approach as helping us in creating features for our model. Therefore it is a good rule to:

**Start with rules and then use these rules as features.**

features that are take 1 of two values. ex: either true or false / 1 or 0, are called **binary features**

We **encode** the feature values of true and false as 1 and 0.

features + target are then used to train a model. 

**Training a model is also called fitting.**
