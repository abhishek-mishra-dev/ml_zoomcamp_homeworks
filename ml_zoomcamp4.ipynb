## CRISP-DM

**What will be covered:**

CRISP-DM - cross industries standard processing data mining

from problem understanding to deployment

taking example of spam detection

**Business understanding:** 

Identify the problem, how to find the solution.

An important question to ask at this stage is: **Do we really need ML to solve it?**

#### IMPORTANT!!!

There has to be metric which tells us if we solved the problem or not and by how much

**Data understanding:**

**Data preparation:**

Transform the data so that it can be put into ML algorithm


- Clean the data (remove the noise)
- Build the pipelines (raw data -> transformation -> clean data (ex: matrix of numbers in 0s and 1s))
- convert into tabular form

**Modelling:** 

actual ML happens here!

Which model to choose?

- Linear regression

- Decision tree

- Neural network

- and many others

Sometimes, we may go back to data preparation,
- add new features
- fix data issues

**Evaluation:**

is the goal reached?

- evaluate the model on some test data

**Evaluation + deployment**

- evaluate the model on live/ real users (maybe 5% of all users of the service can be tested upon)
- it means deploy the model and then evaluate it


**Iteration**

- Start simple 
- learn from feedback
- improve
