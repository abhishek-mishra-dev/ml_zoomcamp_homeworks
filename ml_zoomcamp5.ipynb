## Model Selection

- logistic regression
- decision tree
- neural network
- or many others

If we develop a model in July and then run it on data from future, lets say august dataset, we get our y values. But at present in july, how can we test our model then? So what is usually done is, from the dataset we already have, we divide the dataset into 80% and 20%. The 80% is used for **training** the model and the remaining 20% is not used for training, but only used for **validation**.

lets say, we find one of the models from LR, RF, XGBOOST, NN to be the best one with 80% accuracy greater than all others, ex:NN.

how can we make sure that the model didnt just get particularly **lucky** with the validation data set and is infact the best model? 

solution: We break the dataset into 3 parts instead of 2.
1. training dataset
2. validation dataset
3. test dataset

Once the training, validation and testing is done on the dataset for a particular model, we combine the training + validation data to make a bigger training dataset and this is then run on test data and it gives better results because the training data is larger now.
