## ML application 1 - 

### Predicting the price of a car on car selling and buying website.

As owners of the website, how do you help a person to sell his car with the right price tag, i.e. how do you suggest a good price to him? Because if he guesses a price himself he may estimate it too high and nobody will buy the car or he may price it too low and lose money since the car was worth more money. How can we suggest him right price / price range?


If you go to a dealership, you would find that the expert can predict a price for your car accurately, it is because he has been seeing different cars with different date of manufacture, model, company, no. of seats, mileage and so on and he knows what that type of car usually costs. So from data, he has found patterns in the data. 

An ML model does exactly what the expert did. Learn from the data about model,mileage,corresponding price,etc and extract patterns from the data.

date of manufacture, model, company, no. of seats, mileage  - "features" (what we know about cars)

price -  "target" (what we want to predict)



features + target --> train a model. 

then use the model and new features data to predict a price.

NOTE: price for every individual car may not be correct, but on average the predictions are correct.
